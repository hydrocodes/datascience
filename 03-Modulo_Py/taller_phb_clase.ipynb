```python
def evp_hs():
  """
  Estimacion evapotranspiracion Hargreaves-Samani
  """
  tx = float(input('Temp max (ºC): '))
  tn = float(input('Temp min (ºC): '))
  tp = np.average([tx, tn])
  ra = float(input('Rad ext (mm/day): '))
  etp = 0.0023*ra*(tp+17.8)*(tx-tn)**0.5
  etp = np.around(etp, 2)
  return(etp)
```

```python
evp_hs()
```

## Visualización: Precipitaciones extremas

```python
pcp_max = pcp_canete['X110783'].max()
pcp_q99 = pcp_canete['X110783'].quantile(0.99)
pcp_q95 = pcp_canete['X110783'].quantile(0.95)
pcp_q90 = pcp_canete['X110783'].quantile(0.90)
idx = pcp_canete.index.values
```

```python
pcp_canete['X110783'].plot(kind='line', figsize=(12,4), linewidth=0.8, alpha=1).set_title('Estacion X110783')
plt.fill_between(idx, pcp_q99, pcp_max, facecolor='red', alpha=0.3)
plt.fill_between(idx, pcp_q95, pcp_q99, facecolor='orange', alpha=0.3)
plt.fill_between(idx, pcp_q90, pcp_q95, facecolor='gold', alpha=0.3)
plt.ylabel('Precipitacion (mm/dia)')
plt.xlabel('Tiempo')
plt.grid(alpha = 0.25)
```
